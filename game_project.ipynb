```python
"""
This code represents a simple text-based game in which the player assumes 
the role of a king who needs to make strategic decisions during a war. 
The player wins the war if their total score reaches or exceeds 10 at any 
point in the game. However, the score can decrease or increase based on 
the player's choices and random events. If the score falls below 10, 
the player loses the war.

"""

# Import required modules
import time
import random

# This function prints a message and then pauses for 2 seconds.
def print_pause(x):
    print(x)
    time.sleep(2)

# This function prompts the user to enter a choice and validates the input. 
# It keeps asking for input until the user enters 1 or 2.
def user_choice(message):
    while True:
        choice = int(input(message))
        if choice != 1 and choice != 2:
            print_pause("Invalid input. Please try again, choose: 1 or 2")
        else:
            return choice

# This function generates a random event and modifies the total score accordingly. 
# The score increases if the event is positive and decreases if it's negative.
def random_event(total_score):
    events = ['Your scouts found a hidden cache of resources. Your score increases by 10', 
              'Your troops have been hit by a disease. Your score decreases by 10',
              'A spy in your ranks has been discovered. Your score decreases by 10.',
              'A neighboring kingdom sent you aid. Your score increases by 10.',
              'Enemy forces launched a surprise attack, causing considerable damage. Your score decreases by 10.',
              'Your soldiers have found an advantageous position. Your score increases by 10.']

    event = random.choice(events)

    if 'increases' in event:
        total_score += 10
    else:
        total_score -= 10
    
    print_pause(event)
    return total_score

# This is the main game function. It guides the player through the game.
def game():
    print_pause("You are the king, and you sit in your palace.")
    print_pause("Suddenly, you know that there is a war on the kingdom, and you must take dissection to protect your kingdom.")
    print_pause("There are long and huge walls surrounding the kingdom.")

    # The arrangement of the player's military units is chosen randomly.
    units = [
        'Archery unit in the front, artillery at the end, infantry and '
        'cavalry in the middle.',
        'Cavalry in the front, archery behind the cavalry, and cavalry '
        'then artillery at the end.',
        'Archery in the front, cavalry behind the archery, and artillery '
        'then infantry at the end.'
        ]

    arrange_units = random.choice(units)
    print_pause(f'Arrange of your units is {arrange_units}')
    print_pause("The enemy has huge catapults.")

    total_score = 10
    print_pause('Total score = {}'.format(total_score))

    def score():
        nonlocal total_score
        print_pause('Total score = {}'.format(total_score))
    
    print_pause("Enter 1 if you want to open the walls of the kingdom and fight face to face")
    print_pause("Enter 2 if you want confrontation over the walls through arrows")
    
    The_choice = user_choice("Enter your choice (1 or 2): ")

    # A random event is triggered which affects the total score.
    total_score = random_event(total_score)
    score()

    # These functions represent the two main strategies the player can choose to fight the enemy.
    def open_walls():
        nonlocal total_score
        print("Your armies faced the enemy armies, but the enemy armies were stronger")
        total_score += 5
        score()

        The_choice1 = user_choice(
            "If you want to step back and close the walls, enter 1."
            " and if you want to complete the war, enter 2: "
            )

        # These functions represent the outcomes of the player's choice in the open_walls strategy.
        def step_back():
            nonlocal total_score
            total_score += 5
            if total_score >= 10:  
                print("You are able to win the war")
            else:
                print("Unfortunately, you don't have enough resources to continue the fight. You lost the war.")
                
        def complete():
            nonlocal total_score
            total_score -= 5
            if total_score >= 10:  
                print("You are able to win the war")
            else:
                print("Unfortunately, you don't have enough resources to continue the fight. You lost the war.")

        if The_choice1 == 1:
            step_back()
            score()
        elif The_choice1 == 2:
            complete()
            score()
        else:
            print("Try again") 

    def arrows():
        nonlocal total_score
        print("The guards who used arrows killed a large number of enemy armies, but the catapults destroyed the walls.")
        total_score += 5
        score()

        The_choice2 = user_choice(
            "If you want to complete the war from behind the rubble, enter 1. "
            "and if you want to meet the enemy face to face enter 2 : "
            )

        # These functions represent the outcomes of the player's choice in the arrows strategy.
        def rubble():
            nonlocal total_score
            total_score += 5
            if total_score >= 10:
                print("You are able to win the war")
            else:
                print("Unfortunately, you don't have enough resources to continue the fight. You lost the war.")

        def meet_enemy():
            nonlocal total_score
            total_score -= 5
            if total_score >= 10:
                print("You are able to win the war")
            else:
                print("Unfortunately, you don't have enough resources to continue the fight. You lost the war.")

        if The_choice2 == 1:
            rubble()
            
            score()
        elif The_choice2 == 2:
            meet_enemy()
            
            score()
        else:
            print("Try again") 

    if The_choice == 1:
        open_walls()
    elif The_choice == 2:
        arrows()
    else:
        print("Try again, choose: 1 or 2")

# Main game loop. The game will keep repeating as long as the player chooses to play again.
again = 'yes'
while again.lower() == 'yes':
    game()
    again = input("If you want to play again enter yes: ")


# If the player does not want to play again, print a farewell message.
if again.lower() == 'no':
    print("Thanks for playing! We hope to see you again soon.")
```
